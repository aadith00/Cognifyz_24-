### Task 1 : Top Cuisines

```python
import matplotlib.pyplot as plt

cuisines = df['Cuisines']

# Get the value counts of cuisines
cuisine_counts = cuisines.value_counts()

# Select the top 3 cuisines
top_3_cuisines = cuisine_counts.head(10)

# Plot a bar chart for the top 3 cuisines
plt.figure(figsize=(8, 6))
top_3_cuisines.plot(kind='bar', color='lightgreen')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.title('Top 10 Cuisines')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the above graph, it can be concluded that the top three cuisines in the dataset are North Indian, Chinese and Mughlai.

```python
# Searching for each cuisines in the column and taking the count
north = 'North Indian'
north_count = (df['Cuisines'] == north).sum()
north_perc = north_count / len(df['Cuisines'])          # Percentage of North Indians cuisines in the dataset

chinese = 'Chinese'
chinese_count = (df['Cuisines'] == chinese).sum()
chinese_perc = chinese_count / len(df['Cuisines'])      # Percentage of Chinese cuisines in the dataset

mugh = 'Mughlai'
mugh_count = (df['Cuisines'] == mugh).sum()
mugh_perc = mugh_count / len(df['Cuisines'])            # Percentage of Mughlai cuisines in the dataset

cuisine_perc = {
    'North Indian': north_perc,
    'Chinese': chinese_perc,
    'Mughlai': mugh_perc
}

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(cuisine_perc.values(), labels=cuisine_perc.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Distribution of top 3 Cuisines in Restaurants')
plt.axis('equal')
plt.show()
```

It can be observed that North Indian cuisines are the most available (in 67.2% of restaurants) followed by Chinese (25.4%) and then Mughlai (7.4%).

### Task 2 : City Analysis

```python
# Getting the number of times each city is repeating in the dataset
df['City'].value_counts().head()
```

It can be seen that New Delhi is repeated the most in the City column which means that it has most number of restaurants.

```python
city_avg = df.groupby('City')['Aggregate rating'].mean().sort_values()
city_avg
```

```python
plt.figure(figsize=(8, 6))
city_avg.tail().plot(kind='bar', color='lightblue')
plt.xlabel('Cities')
plt.ylabel('Average Rating')
plt.title('Average Rating for last 5 cities')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the above graph it can be inferred that Inner City has the highest average rating (4.9) among all other cities.

### Task 3 : Price Range Distribution

```python
# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Price range'], bins=range(1, 7), color='pink', edgecolor='black')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.xticks(range(1, 7)) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The price range distribution of restaurants is shown in the above graph.

```python
price_range_counts = df['Price range'].value_counts().sort_index()

# Calculate the total number of restaurants
total_restaurants = len(df)

# Calculate the percentage of restaurants in each price range category
percentages = (price_range_counts / total_restaurants) * 100

# Combine counts and percentages into a DataFrame
price_range_stats = pd.DataFrame({'Count': price_range_counts, 'Percentage': percentages})

# Display the result
print("Number and percentage of restaurants in each price range category:")
print(price_range_stats)
```

In the above table the distribution of restaurants in each price range is given.

### Task 4 : Online Delivery

```python
online = (df['Has Online delivery'] == 'Yes').sum()
online_perc = online / len(df) * 100
print("Percentage of restuarants with online delivery is: " + str(online_perc))
```
